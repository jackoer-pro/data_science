First of all let's clean those data!

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set_theme(style="whitegrid")

df = pd.read_csv('forbes_richman.csv', encoding='latin1')

df = df.dropna(subset=['Name', 'Net Worth']).reset_index(drop=True)

df['Net Worth (Billion $)'] = (
    df['Net Worth']
    .str.replace('$', '', regex=False)
    .str.replace(' B', '', regex=False)
    .astype(float)
)

for col in ['Industry', 'Country']:
    df[col] = df[col].str.strip()

df_age = df.dropna(subset=['Age'])

print(" Data Cleaned Successfully!")
df.head()
```

Question 1: Is extreme wealth concentrated in specific industries????

```python
industry_wealth = df.groupby('Industry')['Net Worth (Billion $)'].sum().reset_index().sort_values(by='Net Worth (Billion $)', ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(data=industry_wealth, x='Net Worth (Billion $)', y='Industry', hue='Industry', palette='viridis', legend=False)
plt.title('Question 1: Total Net Worth by Industry')
plt.show()
```

And the answer is YESSS. But SUPRISINGLY GUYS!!! We chose the right discipline to work on. 
-The bar chart shows that technology and Fashion & Retail are the dominant forces, holding the largest portion of the world's total wealth. While many industries have billionaires, these two are the "hubs" of global capital.

```python
Question 2: Is extreme wealth mostly accumulated later in life????
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df_age['Age'], bins=30, kde=True, color='teal')
plt.axvline(df_age['Age'].median(), color='red', linestyle='--', label=f'Median Age: {df_age["Age"].median():.0f}')
plt.title('Question 2: Age Distribution of Billionaires')
plt.legend()
plt.show()
```

 And the answer is Yesss again.
Whats up, guys! Heres some relief: I guarantee a lot of you stress about getting rich early in life. But the median age of a billionaire is 64. The distribution shows that extreme wealth rarely happens in your 20s or 30s. It usually peaks between 55 and 75, which just proves that building wealth is a game of patience and long-term scaling.


```python
Question 3: Do some industries produce wealth at a younger age????
```

```python
order = df_age.groupby('Industry')['Age'].median().sort_values().index

plt.figure(figsize=(12, 8))
sns.boxplot(data=df_age, x='Age', y='Industry', order=order, hue='Industry', palette='coolwarm', legend=False)
plt.title('Question 3: Age Distribution Across Industries')
plt.show()
```

Tadaa! Technology again.
    -Technology and Finance have lower median ages and more "outlier" young billionaires. Traditional industries like Metals & Mining or Construction tend to have much higher median ages, as they usually require physical infrastructure and decades of asset accumulation.

```python
Question 4: Is extreme wealth geographically clustered????
```

```python
top_countries = df['Country'].value_counts().head(10).reset_index()
top_countries.columns = ['Country', 'Count']
plt.figure(figsize=(12, 6))
sns.barplot(data=top_countries, x='Count', y='Country', hue='Country', palette='plasma', legend=False)
plt.title('Question 4: Top 10 Countries with the Most Billionaires')
plt.show()
```

-Extreme wealth is highly clustered. The United States and China are far ahead of the rest of the world in terms of billionaires and high-net-worth individuals. If youre looking to enter the world of extreme wealth, these are the two main clusters of opportunityplaces where innovation, business infrastructure, and investment ecosystems are strongest. This concentration is also why many wealthy families in Asian countries invest heavily in sending their children to study abroad.
